```python
import numpy as np
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Flatten, Conv2D, MaxPool2D, Dense, ZeroPadding2D, Dropout, Input
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import fashion_mnist
import os
import gzip
import matplotlib.pyplot as plt
```

```python
def load_mnist(path, kind='train'):
    """Load MNIST data from `path`"""
    labels_path = os.path.join(path,
                               '%s-labels-idx1-ubyte.gz'
                               % kind)
    images_path = os.path.join(path,
                               '%s-images-idx3-ubyte.gz'
                               % kind)

    with gzip.open(labels_path, 'rb') as lbpath:
        labels = np.frombuffer(lbpath.read(), dtype=np.uint8,
                               offset=8)

    with gzip.open(images_path, 'rb') as imgpath:
        images = np.frombuffer(imgpath.read(), dtype=np.uint8,
                               offset=16).reshape(len(labels), 784)

    return images, labels
```

```python
X_train, y_train = load_mnist('fashionmnist', kind='train')
X_test, y_test = load_mnist('fashionmnist', kind='t10k')
```

```python
X_test.shape
```

```python
X_train = X_train.reshape(X_train.shape[0], 28, 28)
X_test = X_test.reshape(X_test.shape[0], 28, 28)
```

```python
X_train.shape
```

```python
X_train = np.expand_dims(X_train, axis=3)
X_test = np.expand_dims(X_test, axis=3)
```

```python
X_train.shape
```

```python
X_train[1000]
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
input_tensor = Input(shape=(28, 28, 1))
x = ZeroPadding2D(padding=(1, 1))(input_tensor)
x = Conv2D(32, 3, activation='relu')(x)
x = Dropout(0.2)(x)
x = MaxPool2D(pool_size=2, strides=2)(x)
x = ZeroPadding2D(padding=(1, 1))(x)
x = Conv2D(64, 3, activation='relu')(x)
x = Dropout(0.2)(x)
x = MaxPool2D(pool_size=2, strides=2)(x)
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
output_tensor = Dense(10, activation='softmax')(x)

model = Model(inputs=input_tensor, outputs=output_tensor)
model.summary()
```

```python
model.compile('adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
y_train_ohe = to_categorical(y_train)
y_test_ohe = to_categorical(y_test)
```

```python
history = model.fit(X_train, y_train_ohe, epochs=5, validation_data=(X_test, y_test_ohe), batch_size=300)
```

```python
def plot_feature_maps(feature_maps, layer_name, output_dir='feature_maps'):
    num_filters = feature_maps.shape[-1]
    size = feature_maps.shape[1]

    grid_size = int(np.ceil(np.sqrt(num_filters)))

    fig, axes = plt.subplots(grid_size, grid_size, figsize=(20, 20))
    axes = axes.flatten()

    for i in range(num_filters):
        ax = axes[i]
        x = feature_maps[0, :, :, i]
        x -= x.mean()
        x /= (x.std() + 1e-5)
        x *= 64
        x += 128
        x = np.clip(x, 0, 255).astype('uint8')
        ax.imshow(x, aspect='auto', cmap='viridis')
        ax.axis('off')

    for i in range(num_filters, len(axes)):
        axes[i].axis('off')

    plt.suptitle(layer_name)
    plt.show()

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    for i in range(num_filters):
        np.save(os.path.join(output_dir, f'{layer_name}_filter_{i}.npy'), feature_maps[0, :, :, i])
```

```python
layer_names = ['input'] + [layer.name for layer in model.layers if 'conv' in layer.name or 'pool' in layer.name]
for layer_name, layer_activation in zip(layer_names, activations):
    if layer_name == 'input':
        plt.imshow(layer_activation[0, :, :, 0], cmap='viridis')
        plt.title('Input Layer')
        plt.axis('off')
        plt.show()
        np.save(os.path.join('feature_maps', 'input_layer.npy'), layer_activation[0, :, :, 0])
    else:
        plot_feature_maps(layer_activation, layer_name)
```
